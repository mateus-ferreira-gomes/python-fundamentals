#### df dataset

```python
# Importação das bibliotecas principais

import pandas as pd  # manipulação de dados (DataFrame)
import numpy as np   # operações numéricas
import seaborn as sns  # visualização de dados (gráficos)
import matplotlib.pyplot as plt  # gráficos
from matplotlib import rc

# Para exibir gráficos diretamente no notebook
%matplotlib inline
```

```python
# Conectar ao Google Drive (onde está o dataset)

from google.colab import drive
drive.mount('/content/drive')
```

```python
# Carregando o dataset de vinhos

df = pd.read_csv('/content/drive/MyDrive/Dataset/wine_quality_dataset.csv')
```

```python
# para ter uma primeira visualização do banco de dados carregado
df
```

```python
# Número de linhas (observações) e colunas (variáveis)
df.shape
```

```python
# Informações gerais (tipos de dados e valores não nulos)
df.info()
```

```python
# Verificar valores nulos por coluna
df.isnull().sum()
```

```python
# Verificar linhas duplicadas
df.duplicated().sum()
```

```python
# Ver nomes das colunas

df.columns
```

```python
# Renomeando as colunas para português

df.columns = [
    'acidez fixa', 'acidez volátil', 'ácido cítrico', 'açúcar residual',
    'cloretos', 'dióxido de enxofre livre', 'dióxido de enxofre total',
    'densidade', 'pH', 'sulfatos', 'álcool', 'qualidade', 'tipo'
]

# Conferir nomes atualizados
df.columns
```

```python
# Estatísticas básicas do dataset

df.describe()
```

```python
df.columns
```

```python

# Relação entre acidez fixa e qualidade dos vinhos

plt.figure(figsize=(8, 6))
sns.scatterplot(x='acidez fixa', y='qualidade', hue='tipo', data=df)

plt.title('Análise: Acidez Fixa vs Qualidade do Vinho')
plt.xlabel('Acidez Fixa')
plt.ylabel('Qualidade')

plt.show()
```

```python
# Distribuição da qualidade por tipo de vinho
plt.figure(figsize=(8, 6))
sns.boxplot(x='tipo', y='qualidade', data=df)

plt.title('Distribuição da Qualidade por Tipo de Vinho')

plt.show()
```

```python
# Comparação do teor alcoólico por tipo
plt.figure(figsize=(8, 6))
sns.boxplot(x='tipo', y='álcool', data=df)

plt.title('Teor Alcoólico por Tipo de Vinho')

plt.show()
```

```python
# Matriz de correlação
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')

plt.title('Correlação entre Variáveis')

plt.show()
```

```python
# Distribuição das variáveis
df.hist(figsize=(14, 10))
plt.suptitle('Distribuição das Variáveis')
plt.show()
```
